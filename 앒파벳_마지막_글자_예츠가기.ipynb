```python
import numpy as np
data =np.array([
    [p,r,i,c,e], [s,h,e,e,p], [b,i,t,e,r], [b,l,a,n,d],[b,o,g,g,y],[a,p,p,l,e], [h,e,l,l,o], [c,r,o,w,n], [r,o,y,a,l], [q,u,e,e,n]]
)
```

```python
#시계열 데이터 : 데이터가 순서를 가지고 있다.
#sample.timestep, feature
#sample(10) 가지고 있는 데이터의 개수
#timestep(하나의 데이터가 가지고 있는 순서)
#feature(26): 하나의 timestep이 표현할 수 있는 수
data.shape
X= data[:, :-1]
y= data[:, -1]
X.shape, y.shape
```

```python
# 딥러닝 모델
# MLP, CNN, TOLO > 여러 데이터를 학습할 수 있으나
#                > 순서에 따라 의미가 달라지는 데이터는 학습은 됨, 결과가 안나옴
#                > 순서를 상관하지 않고 학습
# 시계열 딥러닝 모델
#                > 순서에 따라 의미가 달라지는 데이터는 학습, 결과 잘나옴
#                > 순서를 기억한 채로 학습
```

```python
# 모델링
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN

model = Sequential()

# input_shape = 입력 데이터(문제)의 크기(개수는 상관하지 않고 데이터중 1개의 크기를 의미한다.)
# activation = 'tanh' :임팩트 있는 값들은 기억하나 아닌것들은 잊는 성질
model.add(SimpleRNN(units = 30, input_shape= X.shape[1:] , activation ='tanh'))
#units 알파벳의 갯수
model.add(Dense(units = 26, activation = 'softmax'))


#학습 방법 설정
model.compile(
    loss= 'categorical_crossentropy',
    optimizer = 'adam',
    metrics =['accuracy']
)
# 모델 학습
model.fit(X, y ,epochs = 300)
```
